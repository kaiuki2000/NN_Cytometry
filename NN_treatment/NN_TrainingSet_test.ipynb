```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import csv
# Helper function to store data set values in csv file:
def addToFile(file, data):
    with open(file, 'a', newline='') as f:
        writer = csv.writer(f)
        writer.writerows(data)
```

```python
# Helper function for later:
def check_list(list, pos):
    if len(list) > pos:
        return True
    else:
        return False
```

```python
%matplotlib inline

##################################
# Data used later (parameters):  #
# Sigma for noise:               #
sigma = 2.75e-6                  #
# Peaks:                         #
max_peaks = 6                    #
max_width = 150                  #
# Valleys:                       #
max_valleys = 6                  #
# Number of samples to create:   #
n_samples = 20                   #
# Max peak/valley value: (notar  #
# que como os picos podem somar, #
# vamos obter valores maiores    #
# que max_val)                   #
max_val = 10e-6                  #
##################################

# All of the signals:
signals = []

# Defining 'time' array: 1000 points.
t = np.linspace(0, 999, 1000).tolist()

# Counter:
c=0

for i in range(0, n_samples):
    # Signal filled with 1000 zeros:
    signal = np.zeros(1000).tolist()

    # Smooting slopes!
    n_speed = np.random.randint(20, 250)

    # Peaks:
    n_peaks = np.random.randint(4, max_peaks)
    print(f'Iteration {c}:\nNumber of peaks = {n_peaks}')
    for j in range(0, n_peaks):
        width = np.random.randint(0.1*max_width, max_width)
        value = np.random.uniform(3.5*sigma, max_val) # Was (max_val/2, max_val) before!
        pos = np.random.randint(0, 1000) # This isn't ideal, but whatever.

        for k in range(0, width):
            if(check_list(signal, (pos+k))):
                signal[pos+k] += value

        # Smoothing out peaks:
        for i in range(1, n_speed):
            if(check_list(signal, (pos-i))):
                signal[pos-i] += value*(1-i/n_speed)

        for i in range(0, n_speed-1):
            if(check_list(signal, (pos+width+i))):
                signal[pos+width+i] += value*(1-i/(n_speed-1))


    # Valleys:
    n_valleys = np.random.randint(4, max_valleys)
    print(f'Number of valleys = {n_valleys}\n')
    for j in range(0, n_valleys):
        width = np.random.randint(0.1*max_width, max_width)
        value = np.random.uniform(-3.5*sigma, -max_val) # Was (-max_val/2, -max_val) before!
        pos = np.random.randint(0, 1000) # This isn't ideal, but whatever.

        for k in range(0, width):
            if(check_list(signal, (pos+k))):
                signal[pos+k] += value

        # Smoothing out valleys:
        for i in range(1, n_speed):
            if(check_list(signal, (pos-i))):
                signal[pos-i] += value*(1-i/n_speed)

        for i in range(0, n_speed-1):
            if(check_list(signal, (pos+width+i))):
                signal[pos+width+i] += value*(1-i/(n_speed-1))

    if((n_peaks == 0) & (n_valleys == 0)):
        # Manually computed before.
        noise = np.random.normal(0, sigma, 1000)
        signal = (signal + noise).tolist()
    else:
        # Introducing noise!
        # signal = add_noise(signal)
        noise = np.random.normal(0, sigma, 1000)
        signal = (signal + noise).tolist()

    # Uncomment to see plots of the generated signals.
    # plt.figure(c)
    # plt.clf()
    # plt.grid(True)
    # plt.ylabel(r'Voltage [$\mu V$]')
    # plt.xlabel('Point')
    # plt.plot(t, np.array(signal)*1e6)

    # Incrementing counter:
    c+=1

    
    # Labeling signal as fake ('0'):
    signal.append('0')
    signals.append(signal)
    # signal.clear() # Isto buga?

    if(c%10 == 0):
        addToFile('fakes_final_set.csv', signals) # Uncomment to save to file.
        signals.clear()
        print(f'Iteration {c}: signals = {signals}.')

# np.savetxt('data.csv', signals, delimiter=',')

# Lista:
# np.arrays -> list (t, signal); Done!
# Gravar em partes para csv; Done!
# Momento magn√©tico.
# MVP: %history -g
```
